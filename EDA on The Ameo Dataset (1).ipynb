## EDA on The Ameo Dataset

```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('C:/Users/Vaibhav/Desktop/EDA/data.xlsx - Sheet1.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

```python
# Summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(df.describe())
```

```python

# Information about the dataset
print("\nInformation about the dataset:")
print(df.info())
```

### UNIVARIATE ANALYSIS

#### Salary Analysis

```python
print(df['Salary'].describe())

# Histogram of Salary
plt.figure(figsize=(8, 6))
sns.histplot(df['Salary'], kde=True)
plt.title('Histogram of Salary')
plt.show()

# Box Plot of Salary
plt.figure(figsize=(8, 6))
sns.boxplot(x='Salary', data=df)
plt.title('Box Plot of Salary')
plt.show()

# CDF of Salary
plt.figure(figsize=(8, 6))
sns.ecdfplot(df['Salary'])
plt.title('CDF of Salary')
plt.show()
```

#### College GPA Analysis

```python

print(df['collegeGPA'].describe())

# Histogram of College GPA
plt.figure(figsize=(8, 6))
sns.histplot(df['collegeGPA'], kde=True)
plt.title('Histogram of collegeGPA')
plt.show()

# Box Plot of College GPA
plt.figure(figsize=(8, 6))
sns.boxplot(x='collegeGPA', data=df)
plt.title('Box Plot of collegeGPA')
plt.show()

# CDF of College GPA
plt.figure(figsize=(8, 6))
sns.ecdfplot(df['collegeGPA'])
plt.title('CDF of collegeGPA')
plt.show()
```

#### Salary and Tenure of Working

```python
# Prepration of data
df['DOJ'] = pd.to_datetime(df['DOJ'], errors='coerce')
df['DOL'] = pd.to_datetime(df['DOL'], errors='coerce')

# If DOL is 'present', replace it with the current date
df['DOL'] = df['DOL'].fillna(pd.Timestamp('today'))

# Calculate Tenure as the difference between DOL and DOJ in years
df['Tenure'] = (df['DOL'] - df['DOJ']).dt.days / 365

# Plotting the scatter plot of Tenure vs Salary
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Tenure', y='Salary', data=df)
plt.title('Trend of Salary and Tenure of Working')
plt.xlabel('Tenure (Years)')
plt.ylabel('Salary (INR)')
plt.show()
```
